—Ç–æ—á–Ω–æ—Å—Ç–∏ $|ùë•^* ‚àí ùë•_ùëò | ‚â§ 10^{‚àí3}$ . –ú–µ—Ç–æ–¥ –ù—å—é—Ç–æ–Ω–∞ –∑–∞–ø—É—Å–∫–∞—Ç—å —Å –Ω–∞—á–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏—è, –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ –º–µ—Ç–æ–¥–æ–º –ø—Ä–æ—Å—Ç–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏. –°—Ö–æ–¥–∏—Ç—Å—è –ª–∏ –º–µ—Ç–æ–¥ –ù—å—é—Ç–æ–Ω–∞? –ï—Å–ª–∏ —Å—Ö–æ–¥–∏—Ç—Å—è, —Ç–æ —á–µ—Ä–µ–∑

$G(x)= (x-3)cosx - 1$

$G(0) = -4 < 0, G(\pi + \pi) \approx \pi cos(\pi + \pi) - 1 > 0$ -–Ω–∞ $[0; \pi +\pi]$ –∫–∞–∫ –º–∏–Ω–∏–º—É–º 1 –∫–æ—Ä–µ–Ω—å

–í—ã—á–∏—Å–ª–∏—Ç—å —Å —Ç–æ—á–Ω–æ—Å—Ç—å—é Œµ = $10^{‚Äì3}$ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç–æ—á–µ–∫ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –∫—Ä–∏–≤—ã—Ö

$sin(x + 2) ‚Äì y = 1.5,$

–ú–æ–∂–Ω–æ –ü–æ–ª—É—á–∏—Ç—å –∏–∑ –¥–∞–Ω–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π —Ñ–æ—Ä–º—É–ª—ã –¥–ª—è –ú–µ—Ç–æ–¥–∞ –ü—Ä–æ—Å—Ç–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏:

$y = sin(x+2) - 1.5$

$x = 0.5 - cos(y-2)$

–ü—Ä–æ–∏–∑–≤–µ–¥—ë–º –æ—Ü–µ–Ω–∫—É –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞, –∫–æ—Ç–æ—Ä–æ–º—É –¥–æ–ª–∂–Ω—ã –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—Ç—å x, y:

1) $1 \geq sin(x+2) \geq -1$

$-0.5 \geq sin(x+2) - 1.5 \geq -2.5$

$-0.5 \geq y \geq -2.5$

2) $-2.5 \geq y - 2 \geq -4.5$

    $0 \geq cos(y-2) \geq -1$

   $1 \geq -cos(y-2) \geq 0$

   $1.5 \geq 0.5 - cos(y-2) \geq 0.5$

   $1.5 \geq x \geq 0.5$

–ù–æ–≤–∞—è –æ—Ü–µ–Ω–∫–∞ –Ω–∞ —Ö –ü–æ–∑–≤–æ–ª—è–µ—Ç —É—Ç–æ—á–Ω–∏—Ç—å –æ—Ü–µ–Ω–∫—É –Ω–∞ y:

$-\frac{1}{2} \geq sin(x+2) \geq -1$

$-2 \geq sin(x+2) - 1.5 \geq -2.5$

$-2 \geq y \geq -2.5$

–ü—Ä–æ–≤–µ—Ä–∏–º, —á—Ç–æ –Ω–∞ –¥–∞–Ω–Ω–æ–º –º–Ω–æ–∂–µ—Å—Ç–≤–µ –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω–æ–µ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ $G$ —Å–∂–∏–º–∞—é—â–µ–µ:

$G$ = $\left(\begin{array}{c}
         sin(x+2)-1.5 \\

–ü–æ—Å—Ç—Ä–æ–∏–º –º–∞—Ç—Ä–∏—Ü—É –Ø–∫–æ–±–∏:

D[$G$] = $\left(\begin{array}{cc}
         cos(x+2) & 0 \\

–ü–æ–ª—É—á–µ–Ω–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏ –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç, —á—Ç–æ –ø–æ–¥—á–∏–Ω—ë–Ω–Ω–∞—è –ï–≤–∫–ª–∏–¥–æ–≤–æ–π –Ω–æ—Ä–º–µ –Ω–æ—Ä–º–∞ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–µ –ø—Ä–∏–≤–æ—Å—Ö–æ–¥–∏—Ç 0.8

$10^{-3} = \frac{||e_{n+1}||}{||res||} = 0.8^n \frac{||e_0||}{||res||} \geq  0.8^n \frac{\sqrt{0.5^2 + 0.25 \cdot 0.25}}{\sqrt{4 + 0.25}}$

$log_{0.8}10^{-3}\frac{\sqrt{4.25}}{\sqrt{0.25 \cdot 3}} = 25$

```python
import numpy as np
x = 0.5
y = -2.5
for k in range(25):
    x = 0.5 - np.cos(y-2)
    y = np.sin(x+2) - 1.5
print(x, y)
```
